```python
import requests
import defs
import pandas as pd
```

```python
instrument = 'EUR_USD'
count = 10
granularity = 'H1'
```

```python
session = requests.Session()
url = f"{defs.OANDA_URL}/instruments/{instrument}/candles"
```

```python
params = dict(
    count = count,
    granularity = granularity,
    price = 'MBA'
)
```

```python
response = session.get(url, params=params, headers=defs.SECURE_HEADER)
data = response.json()
```

```python
data.keys()
```

```python
for candles in data['candles']:
    print(candles)
```

```python
prices = ['bid','mid','ask']
ohlc = ['o','h','l','c']
```

```python
our_data = []
for candles in data['candles']:
    if candles['complete'] == False:
        continue
    new_dict = {}
    new_dict['time'] = candles['time']
    new_dict['volume'] = candles['volume']
    
    for price in prices:
        for ohl in ohlc:
            new_dict[f'{price}_{ohl}'] = candles[price][ohl]
    our_data.append(new_dict)
print(our_data[0])
```

```python
candles_df = pd.DataFrame.from_dict(our_data)
candles_df
```

```python
candles_df.to_pickle("EUR_USD_H1.pkl")
```
